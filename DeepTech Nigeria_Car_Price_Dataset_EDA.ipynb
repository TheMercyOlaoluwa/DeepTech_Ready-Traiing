## Removing Outliers

```python
u = df['price'].mean()
sd = df['price'].std()
zscore = np.abs(df['price'] - u)/sd
outliercheck = (zscore > 3) | (zscore < -3)
no_of_outliers = outliercheck.sum()
print(no_of_outliers)
```

```python
def count_outliers(df,column):
    u = df[column].mean()
    sd = df[column].std()
    zscore = np.abs(df[column] - u)/sd
    outliercheck = (zscore > 3) | (zscore < -3)
    no_of_outliers = outliercheck.sum()
    pct_of_outliers = no_of_outliers/len(df) * 100
    return(f'{column}: {no_of_outliers} outliers {pct_of_outliers:.2f}%')
```

```python
count_outliers(df,'price')
```

```python
def count_outliersX(df,columns):
    results= {}
    for col in columns:
        u = df[col].mean()
        sd = df[col].std()
        zscore = np.abs(df[col] - u)/sd
        outliercheck = (zscore > 3) 
        no_of_outliers = outliercheck.sum()
        pct_of_outliers = no_of_outliers/len(df) * 100
        results[col] = f'{no_of_outliers} outliers {pct_of_outliers:.2f}%'
    return results
```

```python
count_outliersX(df,['price','Mileage','Engine Size'])
```

## remove outliers

```python
def remove_outliersX(df, cols, threshold = 3):
    mean = df[cols].mean()
    sd = df[cols].std()
    z = np.abs(df[cols] - mean)/sd
    cleaned = (z < threshold).all(axis=1)
    return df[cleaned]
```

```python
len(df)
```

```python
df_cleaned = remove_outliersX(df, ['price','Mileage','Engine Size'])
len(df_cleaned)
```

```python
count_outliersX(df_cleaned,['price','Mileage','Engine Size'])
```
